```python
def tags (x):

    """
    THIS FUNCTION WILL HELP US TO CLEAN 'Tag' COLUMN. 
    FOR EVERYTHING RELATED TO THE Tags... RETURNS THE Tag Category
    """
    dicc_tags = {  
        'Pizza': '.*Pizza.*|.*talian.*|.*asta.*',
        'Pollo' : '.*Pollo.*|.*Chicken.*|.*Alitas.*',
        'Hamburguesa' : '.*Hamburguesa.*|.*hambu.*|.*America.*',
        'Kebab' : '.*Kebab.*',
        'Empanada' : '.*Empanada.*',
        'Dulce' : '.*ostres.*|.*cafe.*|.*apolitana.*|.*dwich.*|.*Cafetería.*|.*Dulce.*',
        'Asiática' : '.*ushi.*|.*ramen.*|.*japon.*|.*asiati.*|.*asiática.*|.*Fried.*|.*Árabe.*|.*Asiática.*',
        'India' : '.*Halal.*|.*India.*',
        'China' : '.*ina.*|.*China.*',
        'Comida rápida' : '.*ar food.*|.*Comida rápida.*',
        'Mediterránea' : '.*Tapas.*|.*roquetas.*|.*pañola.*|.*ritura.*|.*iterranea.*|.*Árabe.*',
        'SouthAmerica' : '.*nezolana.*|.*atinoam.*|.*olombiana.*|.*exica.*|.*oliviana.*|.*Sudam.*|.*peru.*|.*latina.*|.*Mediterránea.*',
        'Carne' : '.*arrilla.*|.*Carne.*',
        'Burritos' : '.*Burrit.*|.*Burritos.*',
        'Poke' : '.*oke.*|.*Poke.*',
        'Ensaladas' : '.*Saluda.*|.*reconf.*|.*vege.*|.*vegana.*|.*comfor.*|.*Ensaladas.*',
        'Gourmet' : '.*speciali.*|.*harina.*|.*queso.*|.*Gourmet.*',
        'Tienda' : '.*lcohol.*|.*cervez.*|.*veryday.*|.*Tienda.*'}
    

    for key,values in dicc_tags.items():

        if re.match(values, x):
            return key 
```

```python
ue['Tag_'] = ue['Tag'].apply(tags)
```

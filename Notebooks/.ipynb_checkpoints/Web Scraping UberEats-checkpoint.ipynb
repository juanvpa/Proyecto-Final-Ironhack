HAREMOS PROGRAMACIÓN DEFENSIVA PARA ASEGURAR QUE TENEMOS EL MISMO NÚMERO DE DATOS EN TODAS LAS LISTAS

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

#ITERAMOS POR LA LISTA DE LOS LINKS PARA SACAR LOS TAGS DE CADA RESTAURANTE
for i in link:
    try:
        #VAMOS A LA URL
        driver.get(i)
        driver.implicitly_wait(2)
        print("Url - Okey")

        try:
            #ACEPTAMOS LAS COOKIES
            driver.find_element_by_css_selector("#cookie-banner > div > div > div.ah.e2 > button.bc.f0.e8.f3.bj.bk.bl.bm.bn.bo.bt.bu.ba.bb").click()
            driver.implicitly_wait(2)
            print("Cookies - Okey")

            #CERRAMOS LA VENTANA EMERGENTE
            driver.find_element_by_xpath('//*[@id="wrapper"]/div[5]/div/div/div[2]/div[2]/button').click()
            driver.implicitly_wait(2)
            print("Ventana emergente - Okey")

        except:

            #UN CLICK MÁS
            try:
                driver.find_element_by_xpath('//*[@id="main-content"]/div[4]/div[1]/div[1]/a').click()
                driver.implicitly_wait(2)
                
            except:
                driver.find_element_by_xpath('//*[@id="main-content"]/div[4]/div/div[1]/p/a').click()
                driver.implicitly_wait(2)
                print("Un click más - Okey")

            try:
                #SACAMOS LA DATA DE CADA TABLA
                tag.append(driver.find_element_by_xpath('//*[@id="wrapper"]/div[5]/div/div/div[2]/div[3]/div[2]/div[2]').text)
                driver.implicitly_wait(2)
                
            except:
                #AÑADIMOS QUE NO TENEMOS DATO
                tag.append("Sin dato")

            try:
                #AÑADIMOS LA DIRECCIÓN
                direcc.append(driver.find_element_by_xpath('//*[@id="wrapper"]/div[5]/div/div/div[2]/div[3]/button[1]/div/div[2]').text)
                driver.implicitly_wait(2)
                print("Congratulations, tenemos todo guarro!")
                
            except:
                #AÑADIMOS QUE NO TENEMOS DATO
                direcc.append("Sin dato")
                
    except:
        tag.append("Sin Tag")
        direcc.append("Sin Dirección")

        driver.quit
```

CONVERTIMOS EN DATAFRAME, LIMPIAMOS Y A JOINEAMOS

```python
df_todoguarro = pd.DataFrame(todoguarro)
df_link = pd.DataFrame(link)
df_cuatro = pd.DataFrame(barrio)
df_tag = pd.DataFrame(tag)
```

LIMPIAMOS EL DF TODOGUARRO (CONTIENE RESTAURANTE, VALORACION, Nº, GASTOS, DELIVERY FEE, ETA, BARRIO Y COMPETIDOR)

```python
uno = []

for i in range(len(df_todoguarro)):
    #itera por cada fila de df
    dicc = {"Restaurantes": None,
        "Valorado": None,
        "Nota": None,
        "Gastos": None,
        "Delivery_Fee": None,
        "ETA": None,
        "Barrio": None,
        "Competidor": None}
    #Regex que hace split por "\n" y "•"
    cuasilimpio = re.split('\n|•',df_todoguarro.iloc[i][0])
    #Lambda que me quita los espacios en blanco " "
    list_dicc = list(filter(lambda x:(x!=" "),cuasilimpio))
    #Actualizar values de las keys de dicc, usando las condiciones de nuestras columnas
    dicc["Restaurantes"]= list_dicc[0]
    for i in list_dicc:
    #itera por cada elemento de list_dicc
        if i.startswith("Uno de los") or i.startswith("Valorado con"):
            dicc["Valorado"] = i
        elif i.startswith("Los gastos de"):
            dicc["Gastos"] = i
        elif i.startswith("Precio de"):
            dicc["Delivery_Fee"] = i
        elif ("min") in i:
            dicc["ETA"] = i
        elif i.startswith ("1.") or i.startswith("2.") or i.startswith("3.") or i.startswith("4.") or i.startswith("5."):
            dicc["Nota"] = i
            
    uno.append(dicc)
```

LIMPIAMOS EL DF TAGS (CONTIENE TAG Y PRECIOS)

```python
dos = []

for i in range(len(df_tag)):
    #itera por cada fila de df
    dicc = {"Tags": None,
        "Precio": None}
    #Regex que hace split por "\n" y "•"
    cuasilimpio = re.split('\n|•',df_tag.iloc[i][0])
    #Lambda que me quita los espacios en blanco " "
    list_dicc = list(filter(lambda x:(x!=" "),cuasilimpio))
    for i in list_dicc:
    #itera por cada elemento de list_dicc
        if ("€") in i:
            dicc["Precio"] = i
        else:
            dicc["Tags"] = ii
            
    dos.append(dicc)
```

```python
df_uno = pd.DataFrame(uno)
df_uno.columns = ["Restaurantes","Valorado","Nota","Gastos","Delivery_Fee","ETA","Barrio","Competidor"]
df_dos = pd.DataFrame(dos)
df_dos.columns = ["Precio","Tags"]
df_tres = pd.DataFrame(direcc)
df_cuatro = pd.DataFrame(barrio)
```

CONCATENAMOS LOS CUATRO DATAFRAMES EN UN ÚNICO Y FINAL QUE NOS QUEDA LIMPIAR

```python
ubereats = pd.concat([df_uno, df_dos, df_tres, df_cuatro], axis=1)
ubereats.drop(['Barrio',"Competidor"], axis=1, inplace=True)
ubereats.columns = ["Restaurante","Valoración","Nota","Gastos","Delivery_Fee","ETA","Tag","Precio","Dirección","Barrio"]
ubereats.head(3)
```

```python
ubereats.to_csv('UberEats.csv')
```

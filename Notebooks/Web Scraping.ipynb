```python
driver = webdriver.Chrome(ChromeDriverManager().install())
todoguarro = []
links = []
barrio = []

#VAMOS A LA URL
driver.get("https://www.ubereats.com/es")
driver.implicitly_wait(5)
print("Url - Centro")

#ACEPTAMOS LAS COOKIES
driver.find_element_by_css_selector("#cookie-banner > div > div > div.ah.gf > button.bc.gg.gh.gk.bj.bk.bl.bm.bn.bo.bt.bu.ba.bb").click()
driver.implicitly_wait(5)
print("Cookies - Centro")
    
#HACEMOS CLICK EN "CERCA DE MI"
driver.find_element_by_xpath('//*[@id="location-typeahead-home-input"]').click()
driver.implicitly_wait(5)
print("Click dirección - Centro")

#ESCRIBIMOS LA DIRECCIÓN
driver.find_element_by_xpath('//*[@id="location-typeahead-home-input"]').send_keys("Plaza Mayor, Madrid")
driver.implicitly_wait(5)
print("Dirección - Centro")

#CONFIRMAMOS LA DIRECCIÓN RECONOCIDA
driver.find_element_by_xpath('//*[@id="location-typeahead-home-item-0"]').click()
driver.implicitly_wait(5)
print("Confirmamos dirección - Centro")

#HACEMOS CLICK PARA DAR UNA CATEGORÍA
driver.find_element_by_xpath('//*[@id="wrapper"]/header/div/div/div[6]/div/div').click()
driver.implicitly_wait(5)
print("Click categoría - Centro")

#ESCRIBIMOS LA CATEGORÍA
driver.find_element_by_xpath('//*[@id="search-suggestions-typeahead-input"]').send_keys("Comida")
driver.implicitly_wait(5)
driver.find_element_by_xpath('//*[@id="search-suggestions-typeahead-input"]').send_keys(Keys.ENTER)
driver.implicitly_wait(5)
print("Categoría - Centro")
print("Estamos cerca - Centro")
driver.implicitly_wait(5)

#ITERAMOS 80 VECES
for number in range (1,80):
    #SACAMOS LA DATA DE CADA TABLA
    todoguarro.append(driver.find_element_by_xpath(f'/html/body/div[1]/div[1]/div/main/div/div/div[2]/div/div[2]/div[{number}]/div/div/div/div[2]').text)
    driver.implicitly_wait(5)
    print("Congratulations, tenemos todo guarro Centro")
    #AÑADIMOS EL BARRIO QUE HEMOS SACADO
    barrio.append("Centro")
    print("Tenemos Centro")
    #AÑADIMOS LOS LINKS EN LA LISTA DE CARA AL PRÓXIMO WEB SCRAPING
    links.append(driver.find_element_by_xpath(f"/html/body/div[1]/div[1]/div/main/div/div/div[2]/div/div[2]/div[{number}]/div/a").get_attribute("href"))
    print("Tenemos link Centro")


#ITERAMOS POR EL DICCIONARIO DE CARA A 1.DARLE UNA DIRECCIÓN Y 2. AÑADIR LA COLUMNA AL DATAFRAME
for b in barrios:

    #NUEVO LUGAR DONDE DAR LA DIRECCIÓN
    driver.find_element_by_xpath('//*[@id="wrapper"]/header/div/div/a[2]').click()
    driver.implicitly_wait(5)

    #NUEVO LUGAR DONDE DAR LA DIRECCIÓN
    driver.find_element_by_xpath('//*[@id="wrapper"]/div[4]/div/div/div[2]/div[3]/div/div[2]/div[1]/div[2]/a').click()
    driver.implicitly_wait(5)

    #NUEVO LUGAR DONDE DAR LA DIRECCIÓN
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div[4]/div/div/div[2]/div[3]/div/div[1]/input').click()
    driver.implicitly_wait(5)

    #ESCRIBIMOS LA DIRECCIÓN
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div[4]/div/div/div[2]/div[3]/div/div[1]/input').send_keys(barrios[b])
    driver.implicitly_wait(5)
    print(f"Dirección - {b}")
            
    #CONFIRMAMOS LA DIRECCIÓN RECONOCIDA
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/div[4]/div/div/div[2]/div[3]/div/ul/li[1]').click()
    driver.implicitly_wait(5)
    print(f"Confirmamos dirección - {b}")

    #ITERAMOS 80 VECES
    for number in range (1,80):
        try:
            #SACAMOS LA DATA DE CADA TABLA
            todoguarro.append(driver.find_element_by_xpath(f'/html/body/div[1]/div[1]/div/main/div/div/div[2]/div/div[2]/div[{number}]/div/div/div/div[2]').text)
            driver.implicitly_wait(5)
            print(f"Congratulations, tenemos todo guarro {b}")
            
        except:
            try:
                #SACAMOS LA DATA DE CADA TABLA
                todoguarro.append(driver.find_element_by_xpath(f'//*[@id="main-content"]/div/div/div[2]/div/div[2]/div[{number}]/div/div/div/div[2]').text)
                driver.implicitly_wait(5)
                print(f"Congratulations, tenemos todo guarro {b}")
            
            except:
                #SACAMOS LA DATA DE CADA TABLA
                todoguarro.append(driver.find_element_by_xpath(f'/html/body/div[1]/div[1]/div/main/div/div/div[2]/div/div[2]/div[{number}]/div/div/div/div[2]').text)
                driver.implicitly_wait(5)
                print(f"Congratulations, tenemos todo guarro {b}")
            
        #AÑADIMOS EL BARRIO QUE HEMOS SACADO
        barrio.append(b)
        print(f"Tenemos {b}")
        #AÑADIMOS LOS LINKS EN LA LISTA DE CARA AL PRÓXIMO WEB SCRAPING
        links.append(driver.find_element_by_xpath(f"/html/body/div[1]/div[1]/div/main/div/div/div[2]/div/div[2]/div[{number}]/div/a").get_attribute("href"))
        print(f"Tenemos link {b}")

        driver.quit
```

```python
uno = []

for i in range(len(df)):
    #itera por cada fila de df
    dicc = {"Restaurantes": None,
        "Valorado": None,
        "Nota": None,
        "Gastos": None,
        "Delivery_Fee": None,
        "ETA": None,
        "Barrio": None,
        "Competidor": None}
    #Regex que hace split por "\n" y "•"
    cuasilimpio = re.split('\n|•',df.iloc[i][0])
    #Lambda que me quita los espacios en blanco " "
    list_dicc = list(filter(lambda x:(x!=" "),cuasilimpio))
    print(list_dicc)
    #Actualizar values de las keys de dicc, usando las condiciones de nuestras columnas
    dicc["Restaurantes"]= list_dicc[0]
    for i in list_dicc:
    #itera por cada elemento de list_dicc
        if i.startswith("Uno de los") or i.startswith("Valorado con"):
            dicc["Valorado"] = i
        elif i.startswith("Los gastos de"):
            dicc["Gastos"] = i
        elif i.startswith("Precio de"):
            dicc["Delivery_Fee"] = i
        elif ("min") in i:
            dicc["ETA"] = i
        elif i.startswith ("1.") or i.startswith("2.") or i.startswith("3.") or i.startswith("4.") or i.startswith("5."):
            dicc["Nota"] = i
            
    uno.append(dicc)
```

```python
df = pd.DataFrame(uno)
df.columns = ["Restaurantes","Valorado","Nota","Gastos","Delivery_Fee","ETA","Barrio","Competidor"]
df.tail()
```
